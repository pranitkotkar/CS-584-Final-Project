Four years' worth of average daily consumption

```python
avg_consumption = elec_consumption.groupby(['Date']).mean()
plt.style.use('seaborn-whitegrid')

# Extracting column names
columns = avg_consumption.columns

# Number of rows and columns for the subplots
nrows = (len(columns) + 1) // 2
ncols = 2 if len(columns) > 1 else 1

# Creating subplots
fig, axs = plt.subplots(nrows, ncols, figsize=(30, 25), squeeze=False)

# Loop to create each subplot
for i in range(nrows):
    for j in range(ncols):
        col_index = i * ncols + j
        if col_index < len(columns):
            axs[i, j].plot(avg_consumption[columns[col_index]])
            axs[i, j].set_title(columns[col_index], fontweight='bold', size=20)
        else:
            # Hide the subplot if it's not needed
            axs[i, j].axis('off')

# Adjust layout to prevent overlap
plt.tight_layout()
```

Monthly average consumption

```python

# Group by month and calculate the average
avg_monthly_consumption = elec_consumption.groupby(elec_consumption['Date'].dt.strftime('%B')).mean()
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
avg_monthly_consumption = avg_monthly_consumption.reindex(months)

plt.style.use('seaborn-whitegrid') 

# Extracting column names
columns = avg_monthly_consumption.columns

# Number of rows and columns for the subplots
nrows = (len(columns) + 1) // 2
ncols = 2 if len(columns) > 1 else 1

# Creating subplots
fig, axs = plt.subplots(nrows, ncols, figsize=(30, 25), squeeze=False)

# Loop to create each subplot
for i in range(nrows):
    for j in range(ncols):
        col_index = i * ncols + j
        if col_index < len(columns):
            axs[i, j].plot(avg_monthly_consumption[columns[col_index]])
            axs[i, j].set_title(columns[col_index], fontweight='bold', size=20)
        else:
            # Hide the subplot if it's not needed
            axs[i, j].axis('off')

# Adjust layout to prevent overlap
plt.tight_layout()
```

```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters()
```

```python
elec_consumption2 = pd.read_csv('household_power_consumption.txt', sep=';', 
                 parse_dates={'dt' : ['Date', 'Time']}, infer_datetime_format=True, 
                 low_memory=False, na_values=['nan','?'], index_col='dt')
```

The weekly average of each day's consumption

```python
days_of_week = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday']
weekly_consumption_day = elec_consumption2.groupby(elec_consumption2.index.day_name()).mean().reindex(days_of_week)

cols = weekly_consumption_day.columns
plt.style.use('seaborn-whitegrid')
# Number of rows and columns for the subplots
nrows = (len(cols) + 1) // 2
ncols = 2 if len(cols) > 1 else 1

# Creating subplots
fig, axs = plt.subplots(nrows, ncols, figsize=(30, 25), squeeze=False)

# Loop to create each subplot
for i in range(nrows):
    for j in range(ncols):
        col_index = i * ncols + j
        if col_index < len(cols):
            axs[i, j].plot(weekly_consumption_day[cols[col_index]])
            axs[i, j].set_title(cols[col_index], fontweight='bold', size=20)
        else:
            # Hide the subplot if it's not needed
            axs[i, j].axis('off')

# Adjust layout to prevent overlap
plt.tight_layout()
```

Average consumption for each hour of the day

```python
daily_consumption = elec_consumption2.resample('H').sum()
daily_consumption.index = daily_consumption.index.time
daily_consumption_hour = daily_consumption.groupby(daily_consumption.index).mean()

cols = daily_consumption_hour.columns
plt.style.use('seaborn-whitegrid')
# Number of rows and columns for the subplots
nrows = (len(cols) + 1) // 2
ncols = 2 if len(cols) > 1 else 1

# Creating subplots
fig, axs = plt.subplots(nrows, ncols, figsize=(30, 25), squeeze=False)

# Loop to create each subplot
for i in range(nrows):
    for j in range(ncols):
        col_index = i * ncols + j
        if col_index < len(cols):
            axs[i, j].plot(daily_consumption_hour[cols[col_index]])
            axs[i, j].set_title(cols[col_index], fontweight='bold', size=20)
        else:
            # Hide the subplot if it's not needed
            axs[i, j].axis('off')

# Adjust layout to prevent overlap
plt.tight_layout()
```
